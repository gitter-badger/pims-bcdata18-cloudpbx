## Some important CDR table fields

Below is a description of most of the fields from the primary `cdr` table.

| column name | description |
|------------|-----------------------------|
| `calldate` | the call start time UTC. |
| `callend`  | the call end time UTC.   |
| `duration` | the seconds from beginning to end of SIP signaling session for a call. |
| `connect_duration` | the seconds from beginning to end of RTP audio media stream for a call. |
| `progress_time` | the seconds needed from SIP signaling session start to receiving an affirmative ring confirmation. |
| `first_rtp_time` | the seconds needed from SIP signaling session start to RTP audio media stream start. |
| `caller` | the sanitized caller number. |
| `caller_domain` | the sanitized caller SIP domain or host. |
| `caller_reverse` | the sanitized caller SIP domain or host reverse. |
| `callername` | the sanitized caller text. |
| `callername_reverse` | the sanitized caller text reversed. |
| `called` | the sanitized called number. |
| `called_domain` | the sanitized called SIP domain or host. |
| `called_reverse` |  the sanitized called SIP domain or host reverse. |
| `sipcallerip` | the IP address of the caller's SIP session. |
| `sipcallerport` | the IP port of the caller's SIP session. |
| `sipcalledip` | the IP address of the called's SIP session. |
| `sipcalledport` | the IP port of the called's SIP session. |
| `whohanged` | who ended to SIP session: caller or called. |
| `lastSIPresponse_id` | the final SIP packet type used at the end of a SIP session. Typically BYE or 200 OK. |
| `dscp` | the ethernet packet ToS/CoS priority tag. |
| `a_payload` | caller's RTP media audio codec format. See linked table. |
| `b_payload` | called's RTP media audio codec format. See linked table. |
| `a_saddr` | the caller's RTP media audio IP address. |
| `b_saddr` | the called's RTP media audio IP address. |
| `a_received` | the number of packets received from the caller's RTP media audio stream. |
| `b_received` | the number of packets received from the called's RTP media audio stream. |
| `a_lost` | the number of packets missing from the caller's RTP media audio stream. |
| `b_lost` | the number of packets missing from the called's RTP media audio stream. |
| `a_ua_id` | the caller's User Agent firmware or software name and version. See UA table. |
| `b_ua_id` | the called's User Agent firmware or software name and version. See UA table. |
| `a_avgjitter_mult10` | the caller's RTP media audio stream average jitter (diff in latency). 10x to avoid decimal. |
| `b_avgjitter_mult10` | the called's RTP media audio stream average jitter (diff in latency). 10x to avoid decimal. |
| `a_maxjitter` | the caller's RTP media audio stream maximum jitter (diff in latency). |
| `b_maxjitter` | the called's RTP media audio stream maximum jitter (diff in latency). |
| `a_sl1` | the number of occurrences where the caller's RTP media audio stream was missing 1 isolated packet. |
| `a_sl2` | the number of occurrences where the caller's RTP media audio stream was missing 2 sequential packets together. |
| `a_sl3` | the number of occurrences where the caller's RTP media audio stream was missing 3 sequential packets together. |
| `a_sl4` | the number of occurrences where the caller's RTP media audio stream was missing 4 sequential packets together. |
| `a_sl5` | the number of occurrences where the caller's RTP media audio stream was missing 5 sequential packets together. |
| `a_sl6` | the number of occurrences where the caller's RTP media audio stream was missing 6 sequential packets together. |
| `a_sl7` | the number of occurrences where the caller's RTP media audio stream was missing 7 sequential packets together. |
| `a_sl8` | the number of occurrences where the caller's RTP media audio stream was missing 8 sequential packets together. |
| `a_sl9` | the number of occurrences where the caller's RTP media audio stream was missing 9 sequential packets together. |
| `a_sl10` | the number of occurrences where the caller's RTP media audio stream was missing 10 sequential packets together. |
| `a_d50` | the number of caller RTP media audio packets which arrived with between 50 to 70 ms of deviation. |
| `a_d70` | the number of caller RTP media audio packets which arrived with between 70 to 90 ms of deviation. |
| `a_d90` | the number of caller RTP media audio packets which arrived with between 90 to 120 ms of deviation. |
| `a_d120` | the number of caller RTP media audio packets which arrived with between 120 to 150 ms of deviation. |
| `a_d150` | the number of caller RTP media audio packets which arrived with between 150 to 200 ms of deviation. |
| `a_d200` | the number of caller RTP media audio packets which arrived with between 200 to 300 ms of deviation. |
| `a_d300` | the number of caller RTP media audio packets which arrived with > 300 ms of deviation. |
| `b_sl1` | the number of occurrences where the called's RTP media audio stream was missing 1 isolated packet. |
| `b_sl2` | the number of occurrences where the called's RTP media audio stream was missing 2 sequential packets together. |
| `b_sl3` | the number of occurrences where the called's RTP media audio stream was missing 3 sequential packets together. |
| `b_sl4` | the number of occurrences where the called's RTP media audio stream was missing 4 sequential packets together. |
| `b_sl5` | the number of occurrences where the called's RTP media audio stream was missing 5 sequential packets together. |
| `b_sl6` | the number of occurrences where the called's RTP media audio stream was missing 6 sequential packets together. |
| `b_sl7` | the number of occurrences where the called's RTP media audio stream was missing 7 sequential packets together. |
| `b_sl8` | the number of occurrences where the called's RTP media audio stream was missing 8 sequential packets together. |
| `b_sl9` | the number of occurrences where the called's RTP media audio stream was missing 9 sequential packets together. |
| `b_sl10` | the number of occurrences where the called's RTP media audio stream was missing 10 sequential packets together. |
| `b_d50` | the number of called RTP media audio packets which arrived with between 50 to 70 ms of deviation. |
| `b_d70` | the number of called RTP media audio packets which arrived with between 70 to 90 ms of deviation. |
| `b_d90` | the number of called RTP media audio packets which arrived with between 90 to 120 ms of deviation. |

| `b_d150` | the number of called RTP media audio packets which arrived with between 150 to 200 ms of deviation. |
| `b_d200` | the number of called RTP media audio packets which arrived with between 200 to 300 ms of deviation. |
| `b_d300` | the number of called RTP media audio packets which arrived with > 300 ms of deviation. |
| `a_mos_f1_mult10` | a predictive Mean Opinion Score of the caller's RTP media audio call quality using a fixed 50ms jitter (diff latency) buffer. Saved at 10x value to avoid decimal. |
| `a_mos_f2_mult10` | a predictive Mean Opinion Score of the caller's RTP media audio call quality using a fixed 200ms jitter (diff latency) buffer. Saved at 10x value to avoid decimal. |
| `a_mos_adapt_mult10` | a predictive Mean Opinion Score of the caller's RTP media audio call quality using an adaptive maximum 500ms length jitter (diff latency) buffer. Saved at 10x value to avoid decimal. |
| `b_mos_f1_mult10` | a predictive Mean Opinion Score of the called's RTP media audio call quality using a fixed 50ms jitter (diff latency) buffer. Saved at 10x value to avoid decimal. |
| `b_mos_f2_mult10` | a predictive Mean Opinion Score of the called's RTP media audio call quality using a fixed 200ms jitter (diff latency) buffer. Saved at 10x value to avoid decimal. |
| `b_mos_adapt_mult10` | a predictive Mean Opinion Score of the called's RTP media audio call quality using an adaptive maximum 500ms length jitter (diff latency) buffer. Saved at 10x value to avoid decimal. |
| `a_rtcp_loss` | loss of RTCP format report packets from the caller's SIP User Agent |
| `a_rtcp_maxfr` | RTP media audio stream packet percent (fraction) loss RTCP format report as seen by the caller's SIP User Agent receiving the called's RTP media audio stream. The caller's SIP User Agent is acting like a remote sensor and sending summaries results approximately once every 5 seconds |
| `a_rtcp_avgfr_mult10` | an average of the above in 10 packet samples. More investigation required. |

The called's SIP User Agent is acting like a remote sensor and sending summaries results approximately once every 5 seconds. |
| `b_rtcp_avgfr_mult10` | an average of the above in 10 packet samples. More investigation required. |
| `b_rtcp_maxjitter` | RTP media audio stream maximum reported jitter in RTCP format report as seen by the called's SIP User Agent receiving the called's RTP media audio stream. The called's SIP User Agent is acting like a remote sensor and sending summaries results approximately once every 5 seconds. |
| `b_rtcp_avgjitter_mult10` | an average of the above in 10 packet samples. More investigation required. |
| `packet_loss_perc_mult1000` | the percent of combined caller/called RTP media audio packet loss. 1000x to avoid decimal. |
| `a_packet_loss_perc_mult1000` | the percent of caller's RTP media audio packet loss. 1000x to avoid decimal. |
| `b_packet_loss_perc_mult1000` | the percent of called's RTP media audio packet loss. 1000x to avoid decimal. |
| `delay_sum` | the summary of all delayed arrival packets from both caller and called expressed in ms. More investigation required. |
| `a_delay_sum` | the summary of all delayed arrival packets from the caller expressed in ms. More investigation required. |
| `b_delay_sum` | the summary of all delayed arrival packets from the called expressed in ms.. More investigation required. |
| `a_rtp_ptime` | RTP media audio packet sample frequency from the caller. Typically 1 RTP packet every 20ms. |
| `b_rtp_ptime` | RTP media audio packet sample frequency from the called. Typically 1 RTP packet every 20ms. |

```python
pd.options.display.max_columns = 150
plt.rcParams['font.size'] = 16
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['axes.labelsize'] = 18
```

## Import/format

```python
# read dataframe. calldate and callend are the only date columns
df = pd.read_csv(DATA_PATH, parse_dates=['calldate', 'callend'])
df = df[DESCRIBED_COLUMNS]
# remo`ve any columns that are entirely NA.
df.dropna(axis=1, how='all', inplace=True)
# some columns have "sanitized" written the whole way down:
#    remove these.
df = df.loc[:, ~(df == 'sanitized').all(axis=0)]
# ## sighup was removed by the DESCRIBED_COLUMNS subsetting
# XX sighup column is entirely zeros. remove it. 
# XX df.drop(columns=['sighup'], inplace=True) 
# create a version of duration column of type timedelta
df.drop(columns=[col for col in df.columns if 'reverse' in col], inplace=True)
df['duration_td'] = df.callend - df.calldate
```

```python
# a dataframe where the rows have been removed if 
# there are more than 60% NA values in those rows.
df_dens = df.dropna(axis=0, thresh = 79)
```

### Distribution of call duration

```python
cutoff = 1000
duration_cutoff = df.duration.values
print('excluding {} durations equal to zero.'.format((duration_cutoff == 0).sum()))
duration_cutoff = duration_cutoff[duration_cutoff < cutoff]
duration_cutoff = duration_cutoff[duration_cutoff > 0]

plt.subplots(1,2,figsize=(15, 5))
plt.subplot(121)
plt.hist(duration_cutoff, bins=50);
plt.xlabel('duration (sec)')
plt.ylabel('count');
plt.subplot(122)
plt.hist(np.log(duration_cutoff), bins=25);
plt.xlabel('log10(duration)');
```

```python
# remove NA values from connect_duration so we can look at the 
# time delta between duration and connect_duration
connect_duration_dense = df.dropna(subset=['connect_duration'])
delta_connect = (connect_duration_dense.duration - 
                 connect_duration_dense.connect_duration)
```

```python
plt.figure(figsize=(12,4))
plt.hist(delta_connect.values, bins=50)
plt.ylabel('count')
plt.xlabel('duration - connect_duration (sec)');
```
